```python
csvs = glob.glob("*.csv")
csvs.sort()
csvs[:5]
```

```python
csvs_09 = glob.glob("*_09.csv") 
csvs_99 = glob.glob("*_99.csv")
csvs_old = csvs_09 + csvs_99
csvs_old.sort()
csvs_old[:5]
```

```python
new_files = [c for c in csvs if c not in csvs_old]
new_files.sort()
new_files[:5]
```
